```python
#using the Housing csv file we want to implement a liniar regresion algorithm in python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
import pandas as pd 

#Load the dataset
data = pd.read_csv("./database/Housing.csv")
print(data)
#Extract the features for this example of a simple liniar regeresion
# the target is the price and the independent feature is the lotsize
X = data["price"]
Y = data["lotsize"]

#Reshape for the conpatibility with sci kit lear
X = X.to_numpy().reshape(len(X), 1)
Y = Y.to_numpy().reshape(len(Y), 1)

#split the data into the test set and the 
trainnig_set = X[:-250]
trainning_res = Y[:-250]
test_set = X[-250:]
test_res = Y[-250:]

#plot the dots of the features
plt.scatter(test_set, test_res, color="black")
plt.title("Test sample")
plt.xlabel("lotsize")
plt.ylabel("price")

#Train the model
model = linear_model.LinearRegression()
model.fit(trainnig_set, trainning_res)

#plot the results for the testing 
plt.plot(test_set, model.predict(test_set), color="green", linewidth=3)
plt.show()

```
