```python
#how to work with the grediant decent on a sample with onbly one feature
import numpy as np
from numpy import c_
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
#implementing a regression without the gradient decent

#generate the data set
dataset, targetset = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)
targetset = targetset.reshape(-1, 1)
m = dataset.shape[0]

#_c() function concatenates putting them on collumns arrays => nD array 
_2Ddataset = np.c_[np.ones((m, 1)), dataset]
theta = np.array([[2.0], [3.0]])

plt.figure(figsize=(10, 5))
plt.scatter(dataset, targetset, color="blue", label="Actual Data")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression without the Gradient Descent")
plt.plot(dataset, _2Ddataset.dot(theta), color="green", label="initial Line(No GD)")
plt.legend()
plt.show()

#with the GD 
learning_rate = 0.1
n_interations = 100

for _ in range(n_interations):
    #Predictions(dot product between the params )
    pred = _2Ddataset.dot(theta)
    #Computing the gradient
    gradient = (2/m) * _2Ddataset.T.dot(pred - targetset)
    #Updating the params
    theta -= learning_rate * gradient

plt.figure(figsize=(10, 5))
plt.scatter(dataset, targetset, color="blue", label="Actual Data")
plt.plot(dataset, _2Ddataset.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression with Gradient Descent")
plt.legend()
plt.show()
```
