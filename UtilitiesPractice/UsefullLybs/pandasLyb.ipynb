#### Accessing Data With Pandas

```python
# access using head(int) | tail(int)
print(coffee.tail(5))

# getting random rows
print()
# random_state make it deterministic
print(coffee.sample(10), random_state=1)
```

```python
# location attribute where we put the row and the column
# it can get a index, a list of indexes and slices
# divided by comma you can search only for some columns
print(coffee.loc[[0, 1, 2]])

print()
print(coffee.loc[5:8, ['Day', 'Units Sold']])

# iloc() allows for usage of indexes to obtain columns
print()
print(coffee.iloc[[1, 2], [0, 1]])
```

```python
# Using another value for the index
# then the iloc() and loc() will not work as before
coffee.index = coffee["Day"]
coffee.head()
```

```python
# Modifing the values
coffee.loc[1, 'Units Sold'] = 10
coffee
```

```python
# When resetting the index by default the old index is added as a new column to the DataFrame
# to reset the index without adding it to the DataFrame use drop=True as an attribute
coffee = coffee.reset_index(drop=True)
coffee
```

```python
# Obtimized way to get the values
coffee.at[0, 'Units Sold']

# iat = index at
coffee.iat[0, 0]

# Grab columns 
coffee['Units Sold']
# Or for single word columns
coffee.Day
```

```python
# Sorting data (using a list ensures multiple label sorting)
# ascending can be boolean for one sort label, but for many 
# you can use a list of 0 && 1: 0 - descending, 1 - ascending
coffee.sort_values(['Units Sold', 'Coffee Type'], ascending=[0, 1])
```

```python
# iterating with a for - though it losses the performance

for index, row in coffee.iterrows():
    print(index)
    print(row['Units Sold'])
    
```
